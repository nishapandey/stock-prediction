```python
#subplots and legends
days = [1,2,3,4,5]
original = [10,20,25,40,50]
predicted = [5,15,25,44,55]

plt.plot(days,original,'b-o', label="original");
plt.plot(days,predicted,'r-o',label="predicted");
plt.legend();
plt.title("plot with legends")
plt.xlabel('week days')
plt.ylabel('price')

```

```python
plt.subplot(3,1,1);
plt.plot(days,original,'b-o', label="original");
plt.legend();
plt.title("original with legends")
plt.xlabel('week days')
plt.ylabel('price')

plt.subplot(3,1,2);
plt.plot(days,predicted,'r-o',label="predicted");
plt.legend();
plt.title("predicted with legends")
plt.xlabel('week days')
plt.ylabel('price')

#price difference
price_diff = [o-p for o,p in zip(original,predicted)]
plt.subplot(3,1,3);
plt.bar(days,price_diff,color='green',label='price_diff');
plt.legend();
plt.title("price_diff with legends")
plt.xlabel('week days')
plt.ylabel('price')
plt.tight_layout()

#save the images
plt.savefig("eg_images",dpi=500)
```
